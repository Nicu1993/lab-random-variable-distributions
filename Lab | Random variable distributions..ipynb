```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
df
```

```python
#Get the numerical variables from our dataset.
```

```python
numerical = df.select_dtypes(include=[np.number])
```

```python
numerical
```

```python
#Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.
```

```python
for column in numerical.columns:
    plt.figure(figsize=(8, 4))
    
    # Plot a histogram
    sns.histplot(numerical[column], kde=True, bins=30)
    
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.
```

```python
for column in numerical.columns:
    plt.figure(figsize=(8, 4))
    
    # Apply log transformation
    log_transformed = np.log1p(numerical[column])
    
    # Plot a histogram
    sns.histplot(log_transformed, kde=True, bins=30)
    
    plt.title(f'Log-Transformed Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
